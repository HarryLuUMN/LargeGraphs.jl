# LargeGraphs staged pipeline demo

This notebook shows how to split layout and rendering so one layout result can be reused across multiple renders or exports, including large-graph-friendly rendering profiles.


```julia
using Pkg
Pkg.activate(isdir(joinpath(pwd(), "src")) ? pwd() : joinpath(pwd(), ".."))
Pkg.instantiate()

using LargeGraphs
```

```julia
nodes = [
    (id="a", label="Alpha", color="#2563eb"),
    (id="b", label="Beta", color="#059669"),
    (id="c", label="Gamma", color="#d97706"),
    (id="d", label="Delta", color="#dc2626"),
]

edges = [
    (source="a", target="b"),
    (source="b", target="c"),
    (source="c", target="d"),
    (source="a", target="d"),
]
```

```julia
layouted = layout_graph(
    nodes,
    edges;
    layout=:force_directed,
    algorithm=:forceatlas2,
    iterations=60,
    seed=7,
)

layouted.nodes
```

```julia
viz = assemble_graph(
    layouted;
    profile=:large,
    config=SigmaConfig(height="420px", background="#f8fafc"),
)

viz
```

```julia
savehtml("staged-pipeline-demo.html", viz)
```
