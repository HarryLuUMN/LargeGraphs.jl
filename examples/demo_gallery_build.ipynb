# LargeGraphs gallery build walkthrough

This notebook shows the standalone gallery workflow. The gallery is kept separate from the `Documenter.jl` docs build and writes browseable static output under `gallery/build/`.

```julia
using Pkg

repo_root = isdir(joinpath(pwd(), "gallery")) ? pwd() : normpath(joinpath(pwd(), ".."))
Pkg.activate(repo_root)
Pkg.instantiate()

using LargeGraphs
```

```julia
include(joinpath(repo_root, "gallery", "scripts", "build_gallery.jl"))
gallery_dir = build_gallery()
gallery_dir
```

```julia
readdir(gallery_dir)
readdir(joinpath(gallery_dir, "cases"))
```

Open `gallery/build/index.html` in a browser or publish the contents of `gallery/build/` with any static host. The tutorial material still lives under `examples/`, while the gallery focuses on polished exported outputs.
