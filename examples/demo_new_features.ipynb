# LargeGraphs new feature walkthrough

This notebook demonstrates the new offline export, auto-profile selection, staged timing API, Graphs.jl attribute bridging helpers, and hierarchy layout support.

```julia
using LargeGraphs
using Graphs
```

```julia
nodes = [(id="n$(i)", label="Node $(i)") for i in 1:600]
edges = [(source="n$(i)", target="n$(i + 1)") for i in 1:599]
recommend_profile(nodes, edges), render(nodes, edges; profile=:auto, height="520px")
```

```julia
timed = profile_pipeline(nodes, edges; layout=:grid, layout_kwargs=(columns=25,), export_path="new-features-offline.html")
timed.timings
```

```julia
savehtml("new-features-sigma.html", nodes, edges; self_contained=false, profile=:auto, height="520px")
savehtml("new-features-offline.html", nodes, edges; self_contained=true, profile=:auto, height="520px")
```

```julia
g = path_graph(6)
vertex_attrs = Dict(
    1 => (label="Start", color="#ef4444", size=2.2, attributes=(group="A",)),
    2 => (name="Middle", weight=1.6),
    6 => (label="End", color="#2563eb", size=2.0),
)
edge_attrs = Dict((1, 2) => (label="seed", color="#94a3b8"), "5=>6" => (label="tail", color="#64748b"))
render(
    g;
    layout=:grid,
    node_mapper=vertex_attribute_mapper(vertex_attrs; id_mapper=v -> "v$(v)"),
    edge_mapper=edge_attribute_mapper(edge_attrs; source_mapper=s -> "v$(s)", target_mapper=t -> "v$(t)"),
    height="420px",
)
```

```julia
hierarchy_nodes = [
    (id="root", label="Root", parent=nothing),
    (id="left", label="Left", parent="root"),
    (id="right", label="Right", parent="root"),
    (id="left-left", label="Left Left", parent="left"),
    (id="left-right", label="Left Right", parent="left"),
    (id="right-left", label="Right Left", parent="right"),
]
render(hierarchy_nodes, EdgeSpec[]; layout=:hierarchy, orientation=:left_right, root="root", height="500px")
```
