# Experimental GPU layout

This notebook shows the new experimental CUDA-backed Fruchterman-Reingold layout path.

```julia
using LargeGraphs

nodes = [(id="n$(i)", label="Node $(i)") for i in 1:24]
edges = vcat([(source="n$(i)", target="n$(i+1)") for i in 1:23], [(source="n1", target="n12"), (source="n6", target="n18"), (source="n3", target="n21")])
```

```julia
viz = render(
    nodes,
    edges;
    layout=:force_directed,
    algorithm=:fruchterman_reingold,
    backend=:gpu,
    iterations=120,
    seed=7,
    profile=:dense,
    height="600px",
)

viz
```

If CUDA is not available on the current machine, LargeGraphs falls back to the CPU Fruchterman-Reingold solver.
