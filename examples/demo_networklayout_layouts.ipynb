# NetworkLayout-backed force layouts

This notebook shows the faster `NetworkLayout.jl` integrations exposed through `LargeGraphs`.

```julia
using Pkg
Pkg.activate(isdir(joinpath(pwd(), "src")) ? pwd() : joinpath(pwd(), ".."))
Pkg.instantiate()

using Graphs
using LargeGraphs
```

```julia
g = barabasi_albert(120, 3; seed=7)
```

```julia
render(
    g;
    layout=:force_directed,
    algorithm=:network_spring,
    iterations=80,
    seed=7,
    height="420px",
)
```

```julia
render(
    g;
    layout=:force_directed,
    algorithm=:sfdp,
    iterations=80,
    seed=7,
    height="420px",
)
```
