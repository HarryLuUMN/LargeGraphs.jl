# LargeGraphs experimental WebGPU runtime

This notebook shows the opt-in `runtime=:webgpu` path. The browser runtime prefers WebGPU, falls back to WebGL when needed, and keeps the existing Julia-side graph API unchanged.

```julia
using Pkg
Pkg.activate(isdir(joinpath(pwd(), "src")) ? pwd() : joinpath(pwd(), ".."))
Pkg.instantiate()

using LargeGraphs
```

```julia
nodes = [
    (id="hub", label="Hub", size=4.2, color="#2563eb"),
    (id="analytics", label="Analytics", size=2.7, color="#059669"),
    (id="search", label="Search", size=2.5, color="#d97706"),
    (id="ops", label="Ops", size=2.4, color="#dc2626"),
    (id="ml", label="ML", size=2.2, color="#7c3aed"),
    (id="cache", label="Cache", size=2.0, color="#0891b2"),
]

edges = [
    (source="hub", target="analytics", size=1.0, color="#94a3b8"),
    (source="hub", target="search", size=1.0, color="#94a3b8"),
    (source="hub", target="ops", size=1.0, color="#94a3b8"),
    (source="hub", target="ml", size=0.9, color="#cbd5e1"),
    (source="hub", target="cache", size=0.9, color="#cbd5e1"),
    (source="analytics", target="ml", size=0.8, color="#cbd5e1"),
    (source="search", target="cache", size=0.8, color="#cbd5e1"),
]

viz = render(
    nodes,
    edges;
    layout=:force_directed,
    algorithm=:sfdp,
    iterations=90,
    seed=11,
    runtime=:webgpu,
    height="560px",
    background="#f8fafc",
    label_density=0.7,
)

display(viz)
```

```julia
output = joinpath(pwd(), "demo_webgpu_runtime.html")
savehtml(output, viz)
output
```
