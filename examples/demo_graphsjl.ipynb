# LargeGraphs with Graphs.jl

This notebook shows the new graph-object input path. The package still supports plain `nodes` and `edges`, but it can now also render `Graphs.jl` graphs directly.


```julia
using Pkg
Pkg.activate(isdir(joinpath(pwd(), "src")) ? pwd() : joinpath(pwd(), ".."))
Pkg.instantiate()

using Graphs
using LargeGraphs
```

## Render a graph object directly

Without any mapping functions, vertex ids are stringified and edges are taken directly from the graph structure.


```julia
g = barabasi_albert(40, 2)
render(g; layout=:force_directed, algorithm=:forceatlas2, iterations=80, seed=4, height="520px", hide_edges_on_move=true)
```

## Add labels and styling with `node_mapper` and `edge_mapper`

These functions let the graph-object path feed the same normalized `NodeSpec` / `EdgeSpec` pipeline used by the plain collection path.


```julia
star = star_graph(10)

render(
    star;
    layout=:circular,
    node_mapper=v -> (
        id="v$v",
        label=v == 1 ? "hub" : "leaf-$v",
        size=v == 1 ? 4.0 : 1.8,
        color=v == 1 ? "#2563eb" : "#059669",
    ),
    edge_mapper=e -> (
        source="v$(src(e))",
        target="v$(dst(e))",
        size=0.7,
        color="#94a3b8",
    ),
    height="520px",
)
```
