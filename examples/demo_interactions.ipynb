# LargeGraphs notebook interactions

This notebook demonstrates hover tooltips, click selection, neighbor highlighting, and Julia-side state updates through `InteractionState`.


```julia
using Pkg
Pkg.activate(isdir(joinpath(pwd(), "src")) ? pwd() : joinpath(pwd(), ".."))
Pkg.instantiate()

using IJulia
using LargeGraphs
```

```julia
nodes = [
    (id="a", label="Alpha", size=3.2, color="#2563eb"),
    (id="b", label="Beta", size=2.4, color="#059669"),
    (id="c", label="Gamma", size=2.2, color="#d97706"),
    (id="d", label="Delta", size=2.0, color="#7c3aed"),
    (id="e", label="Epsilon", size=1.8, color="#db2777"),
]

edges = [
    (source="a", target="b", color="#94a3b8"),
    (source="a", target="c", color="#94a3b8"),
    (source="b", target="d", color="#94a3b8"),
    (source="c", target="d", color="#94a3b8"),
    (source="c", target="e", color="#94a3b8"),
]

state = InteractionState()
nothing
```

## Interact with the graph

Hover nodes to see tooltips. Click a node to select it and highlight its neighborhood. Click the background to clear the selection.


```julia
render(
    nodes,
    edges;
    interaction_state=state,
    layout=:force_directed,
    algorithm=:forceatlas2,
    iterations=80,
    seed=4,
    height="520px",
    hide_edges_on_move=true,
)
```

## Inspect Julia-side state

Run the cells below after interacting with the graph output.


```julia
selected_node(state), selected_neighbors(state), hovered_node(state)
```
